```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 创建x轴和y轴
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3,5,7,6,2,6,10,15])
# 绘制折线图 四个参数：x轴，y轴，颜色，线条宽度。r代表红色，g代表绿色
# plt.plot(x,y,'r')
plt.plot(x,y,'g',lw=5)
```

```python
# 绘制柱状图 x轴 y轴 每个柱状图的宽度比例 透明度 颜色
plt.bar(x,y,0.5,alpha=1,color='b')
plt.show()
```

```python
plt.plot(x,y,'g',lw=5)
plt.bar(x,y,0.5,alpha=1,color='b')
plt.show()
```
