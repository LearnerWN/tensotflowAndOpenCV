```python
'''
1.数据准备：yale开源人脸库
2.准备训练数据以及标签
3.构建cnn
4.检测
'''
import tensorflow as tf
import numpy as np
import scipy.io as sio

# 加载数据
file = open('Yale_64x64.mat','rb')
fileDict = sio.loadmat(file)
# 设置参数
learnRate = 0.005
trainSize = 165
testSize = 64
trainTimes = 20000

# yale文件为字典类型 key为fea value为gnd
trainData = fileDict['fea']
trainLable = fileDict['gnd']
# 将数据进行无序排列
trainData = np.random.permutation(trainData)
trainLable = np.random.permutation(trainLable)
# 从训练数据中抽取出64个作测试
testData = trainData[0:64]
testLabel = trainLable[0:64]
np.random.seed(100)
testData = np.random.permutation(testData)
np.random.seed(100)
testLabel = np.random.permutation(testLabel)
# 更改训练数据维度并进行归一化
trainData = trainData.reshape(trainData.shape[0],64,64,1).astype(np.float32)/255
# yale数据库里有165张图片对应的是15个人
trainLabelNew = np.zeros((165,15))
for i in range(0,trainSize):
    # 1-15 -> 0-14
    j = int(trainLable[i,0]) - 1
    trainLabelNew[i,j] = 1
# 更改测试数据维度并进行归一化
testDataInput = testData.reshape(testData.shape[0],64,64,1).astype(np.float32)/255
testLabelNew = np.zeros((64,15))
dataInput = tf.placeholder(tf.float32,[None,64,64,1])
labelInput = tf.placeholder(tf.float32,[None,15])
# 抽取了64个测试数据
for i in range(0,testSize):
    # 1-15 -> 0-14
    j = int(testLabel[i,0]) - 1
    testLabelNew[i,j] = 1

# 构建cnn
layer1 = tf.layers.conv2d(inputs=dataInput,filters=32,kernel_size=2,strides=1,padding='SAME',activation=tf.nn.relu)
layer1Pool = tf.layers.max_pooling2d(layer1,pool_size=2,strides=2)
layer2 = tf.reshape(layer1Pool,[-1,32*32*32])
layer2Relu = tf.layers.dense(layer2,1024,tf.nn.relu)
output = tf.layers.dense(layer2Relu,15)
loss = tf.losses.softmax_cross_entropy(onehot_labels=labelInput,logits=output)
train = tf.train.GradientDescentOptimizer(learnRate).minimize(loss)
accuracy = tf.metrics.accuracy(labels=tf.arg_max(labelInput,1),predictions=tf.arg_max(output,1))[1]

# 运行模型
init = tf.group(tf.global_variables_initializer(),tf.local_variables_initializer())
with tf.Session() as sess:
    sess.run(init)
    for i in range(0,trainTimes):
        trainDataInput = np.array(trainData)
        trainLableInput = np.array(trainLabelNew)
        sess.run([train,loss],feed_dict={dataInput:trainDataInput,labelInput:trainLableInput})
        acc = sess.run(accuracy,feed_dict={dataInput:testDataInput,labelInput:testLabelNew})
        print('第',i,'次acc:%.2f',acc)
```
